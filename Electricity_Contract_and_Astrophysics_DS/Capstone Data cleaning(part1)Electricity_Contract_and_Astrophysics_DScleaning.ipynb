```python
import pandas as pd
import re
from datetime import datetime

df = pd.read_excel(r"c:\Users\yunis\Downloads\data_cleaning  (1).xlsx", sheet_name="Usage", header=None)

def parse(x):
    t = x.replace("_"," ")
    u = float(re.search(r"(\d+\.\d+)", t).group(1))
    h = re.search(r"(\d{1,2}\s*[APMapm]{2})", t).group(1).upper().replace(" ", "")
    d = re.sub(r"(st|nd|rd|th)", "", re.search(r"\d+(?:st|nd|rd|th)?-[A-Za-z]{3}-\d{4}", t).group(0))
    dt = datetime.combine(datetime.strptime(d, "%d-%b-%Y").date(),
                          datetime.strptime(h, "%I%p").time())
    return pd.Series([dt, h, u])

clean = df[0].apply(parse)
clean.columns = ["datetime", "hour", "usage_kwh"]
clean["weekday"] = clean["datetime"].dt.day_name()

clean.head()
```

```python
#question 5(Based on your historic electricity usage, what would your annual cost of electricitybe under the "Monthly Flex" contract?)
contracts = pd.read_excel(r"c:\Users\yunis\Downloads\data_cleaning  (1).xlsx", sheet_name="Contracts", header=None)
contracts
```

```python
#i have flat rate 0.21 so no flex = 0.21 rows 12-23 i have monthly flex so:
monthly_rates = contracts.iloc[12:24, 5].tolist()
monthly_rates
```

```python
monthly_rates = contracts.iloc[12:24, 5].tolist()

clean["month"] = clean["datetime"].dt.month

monthly_usage = clean.groupby("month")["usage_kwh"].sum()

monthly_flex_cost = sum(
    monthly_usage[m] * monthly_rates[m-1]
    for m in range(1, 13)
)

monthly_flex_cost #So the answer is B
```

```python
no_flex_rate = 0.21

no_flex_cost = clean["usage_kwh"].sum() * no_flex_rate
```

```python
hourly_rates = contracts.iloc[12:36, 9].tolist()

clean["hour_num"] = clean["datetime"].dt.hour

hourly_flex_cost = sum(
    clean.iloc[i]["usage_kwh"] * hourly_rates[clean.iloc[i]["hour_num"]]
    for i in range(len(clean))
)
```

```python
#question 6 (Based on your historic electricity usage, which of the three contracts would pro- duce the lowest annual cost?) Answer is C
costs = {
    "No Flex": no_flex_cost,
    "Monthly Flex": monthly_flex_cost,
    "Hourly Flex": hourly_flex_cost
}

print("Q6 Cheapest plan =", min(costs, key=costs.get))
```
